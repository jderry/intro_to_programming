```python
from algebra_with_sympy import * # imports sympy, as well
E, k, c, C, t, y = symbols('E k c C t y', real=True)
```

<p>"Newton's law of cooling is a physical law which states that the rate of heat loss of a body is directly proportional to the difference in the temperatures between the body and its environment. The law is frequently qualified to include the condition that the temperature difference is small and the nature of heat transfer mechanism remains the same. As such, it is equivalent to a statement that the heat transfer coefficient, which mediates between heat losses and temperature differences, is a constant.
</p><p>"When stated in terms of temperature differences, Newton's law...results in a simple differential equation expressing temperature-difference as a function of time. The solution to that equation describes an exponential decrease of temperature-difference over time."
</p>--- wikipedia.org, "Newton's law of cooling"

<p>The "simple differential equation expressing temperature-difference as a function of time" that we'll use is:</p>
<p style="margin-left: 15px;">$dy/dt = k*(y - E)$, where:</p>
<p style="margin-left: 25px;">
$t =$ time;<br>
$y =$ temperature of cooling mass;<br>
$E =$ temperature of environment; and<br>
$k =$ the proportionality constant.</p>
<p>We wish to find an algebraic equation that allows us to solve for $y$, given $t$. To do that requires us to first solve the differential equation and solving for $k$, given $E$.

<p>we start by giving E a numeric value,
then collecting all the y terms on one side of the equation, and everything else on the other:</p>
<p>$1 / (y - E) \: dy = k \: dt$</p>
<p></p>
<p>* This application of calculus we're using to derive our model was presented by
<br>Professor Bruce H. Edwards in <i>Understanding Calculus</i>, "Lesson 36: Applications of Differential Equations, Newton's Law of Cooling". The Great Courses, 2010.</p>

```python
# the left-hand side (LHS) is with respect to y (temp); whereas the right-hand side (RHS) is with respect to t (time)
eqn = Eqn(1 / (y - E), k); eqn
```

```python
# now set up to integrate both sides:
# show the textbook format of the left-hand side (LHS)
LHS = Integral(eqn.lhs, y); LHS
```

```python
# show the textbook format of the right-hand side (RHS)
RHS = Integral(eqn.rhs, t); RHS
```

```python
# our equation looks like this:
Eqn(LHS, RHS)
```

```python
# now integrate. we start by integrating the LHS
LHS = LHS.doit(); LHS # show the resulting formula
```

```python
RHS = RHS.doit(); RHS # now integrate the RHS and show the resulting formula
```

```python
# our equation now looks like this:
eqn = Eqn(LHS, RHS); eqn
```

```python
# add a constant 'c' to the RHS
RHS = eqn.rhs + c
eqn = Eqn(LHS, RHS); eqn
```

```python
# now we raise both sides to e
eqn = exp(eqn); eqn
```

```python
# now factor the RHS.
# observe that a constant raised to the power of e is also a constant
factor(eqn.rhs)
```

Let C $= e^{c}$. Then:

```python
# the resulting expression replaces c raised to the power of e with C
RHS = C * simplify(eqn.rhs / exp(c)); RHS
```

```python
# our equation now looks like this:
eqn = Eqn(eqn.lhs, RHS); eqn
```

# our model

```python
# at this point, our equation is:
eqn = Eqn(eqn.lhs, eqn.rhs); eqn
```

```python
# Setting LHS = 0 by subtracting the LHS from both sides, our equation becomes:
eqn = eqn - eqn.lhs; eqn
```

### by substituting values for variables in our model and applying a little algebra by hand, 
### we can solve for either time t, given temp of mass, or for temp of mass, given time t

#### For reference, we'll start with the input that Dr Edwards uses in his lecture and ensure our output agrees with his.

```python
# He set his problem up to find how long it will take for his cup of coffee to reach a specific temperature,
# given room temperature, the temperature of the coffee at time 0, and the temperature of the coffee at 10 minutes.
# He measures time in minutes and temperatures in Fahrenheit.
num_vals = {t:0, y:100, E:60} # initial values: at time 0, the temp of the coffee is 100F. room temp is a constant 60F.
eqn.subs(num_vals) + 40       # adding 40 to both sides of the equation, solving for C
```

```python
# plugging 40 in for C and adding 30 to both sides of the equation, 
# at time = 10min and coffee temp = 90F, we get:
num_vals = {t:10, y:90, E:60, C:40}
eqn.subs(num_vals) + 30
```

```python
# let's combine the steps to solve for k
# we don't want to modify our model so we'll save our work to a new equation
eqn1 = log((eqn.subs(num_vals) + 30) / 40) / 10; eqn1
```

```python
# finding k lets us use our model for prediction.
# Dr Edwards calculates that it will take ~24.09 minutes for his coffee to cool to 80F
num_vals = {k:eqn1.lhs, y:80, E:60, C:40}
N(log((eqn.subs(num_vals) + 20) / 40) / eqn1.lhs)
```

#### Our result agrees with Dr Edwards', given the same input data.
### Now we can use our method to create models that solve for time or temp in other situations.

```python
# Setting t=0 and y = temp of mass at t=0, we get the value of C.
num_vals = {t:0, y:94.1, E:25.21}
eqn.subs(num_vals)
```

```python
# Setting t = first reading of temp of mass as it cools, and y = temp at that time,
# we're left with k:
num_vals = {t:120, y:90, E:25.21, C:68.89}
soln = eqn.subs(num_vals); soln
```

## the value of a model is its power to predict.
## we can now solve for time t, given temp of mass y; or for temp of mass y, given time t

```python
# to solve for time t in this model, we first provide value of temp of mass, y, at time t
num_vals = {y:80, E:25.21, C:68.89, k:log(64.79/68.89) / 120}
eqn.subs(num_vals)
```

```python
# we then use algebra to collect all terms but t on the LHS of expr. evaluating the LHS returns time t
log(54.79/68.89) / (log(54.79/68.89) / 120)
```

```python
# to solve for temp of mass y at time t in this model, we provide time t:
num_vals = {t:447.86, E:25.21, C:68.89, k:log(64.79/68.89) / 120}
eqn.subs(num_vals)
```
