```python
# scatter plots and color maps
n = 50
x = np.linspace(0,10, n)
noise = np.random.rand(n) * 5.
y1 = x + noise
```

```python
# the default colormap in matplotlib used to be the same as in matlab. 
# it's called 'jet', and it's infamous for how much it distorts data.
# let's add dimensions to our datapoints...

plt.figure()
areas = np.random.rand(n) * 300
colors = np.random.rand(n)
plt.scatter(x, y1, c=colors, s=areas, cmap='jet')
plt.colorbar()

'''
Q: how many dimensions do the datapoints in this scatterplot now have?
Q: does that number make sense (in the context of a 2D plot)?
now to the colormap. the colormap of our scatterplot is the classic 'jet'.
Q: 'jet' is no longer the default colormap for matlab and matplotlib plots. why?
'''
```

```python
# the current default for matplotlib is now 'viridis'
# it's what you get when you don't specify a cmap argument

plt.figure()
areas = np.random.rand(n) * 300
colors = np.random.rand(n)
plt.scatter(x, y1, c=colors, s=areas, cmap='viridis')
plt.colorbar()
'''
Q: is there a problem with viridis?
'''
```

```python
# grayscale works great for single-dimensional data

figure = plt.figure()
areas = np.random.rand(n) * 300
colors = np.random.rand(n)
plt.scatter(x, y1, c=colors, s=areas, cmap='gray')
plt.colorbar()
plt.show()
```

```python
# image plots

# create some data
e1 = np.random.rand(100)
e2 = np.random.rand(100) * 2
e3 = np.random.rand(100) * 10
e4 = np.random.rand(100) * 100

corrmatrix = np.corrcoef([e1, e2, e3, e4])

plt.imshow(corrmatrix, interpolation='nearest', cmap='rainbow_r')
plt.colorbar()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

home_dir = os.path.expanduser('~')

with open(home_dir + '/datafile/reformatted_Unigene.fa') as inFile:
    LoS = inFile.read().splitlines()
    LoL = []
    for record in LoS:
        ''' after splitting record string,
            get the length of item
            in index position 6,
            and append to LoL
        '''
        LoL.append(len(record.split()[6]))
    LoL = np.array(LoL)
binwidth = 1
plt.figure(figsize=(30,5)) # set height and width of figure
plt.hist(LoL, bins=np.arange(LoL.min(), LoL.max()+binwidth, binwidth))
```
