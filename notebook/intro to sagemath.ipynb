```python
for i in range(5):
    print('%6s %6s %6s' % (i, i^2, i^3)) # classical format printing statement
```

```python
for i in range(5):
    print(f"{i:6} {i^2:6} {i^3:6}") # python format printing statement as of python3.6
```

```python
x, a, b, c = var('x a b c')
solve([a*x^2 + b*x + c == 0],x)
```
